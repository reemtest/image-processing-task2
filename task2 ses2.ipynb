```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def display_img(img):
    new_img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(new_img)
    plt.show()

image=cv2.imread('./downloads/baboon.jpg')
display_img(image)
sigma = 3.0
kernel_size = 5
x, y = np.meshgrid(np.arange(kernel_size) - kernel_size//2, np.arange(kernel_size) - kernel_size//2)
kernel = np.exp(-(x*2 + y*2) / (2 * sigma*2))
kernel /= np.sum(kernel)
print (kernel)
img_avg = cv2.filter2D(image, -1, kernel)

display_img(img_avg)
```
